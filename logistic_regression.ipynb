```python
df = pd.read_csv('/content/house price with multiple factors.csv')
df
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd

# 1. Create a binary target variable
# Let's say if price > 400,000, it's "Expensive" (1), otherwise "Affordable" (0)
df['is_expensive'] = (df['price'] > 400000).astype(int)

# Clean column names by stripping whitespace
df.columns = df.columns.str.strip()

# 2. Define Features (X) and the new Target (y)
X = df[['area', 'room', 'floor']]
y = df['is_expensive']

# 3. Initialize and train the Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X, y)

# 4. Predict the category for a house: 3000 sqft, 3 rooms, 4th floor

new_house_data = pd.DataFrame([[3000, 3, 4]], columns=X.columns)
prediction = log_reg.predict(new_house_data)

print(f"Prediction (1 for Expensive, 0 for Affordable): {prediction[0]}")

# 5. Check the probability
probability = log_reg.predict_proba(new_house_data)
print(f"Probability of being affordable vs expensive: {probability}")
```
