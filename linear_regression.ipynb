```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

uploading dataset from local system  as cvs file

```python
df = pd.read_csv('/content/home price.csv') #path of the file which is being uploaded
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

 Predict price of a home with area = 6700 sqr ft

```python
reg.predict([[6700]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
