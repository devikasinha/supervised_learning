```python
import pandas as pd
import numpy as np

np.random.seed(42)
data = np.random.rand(286, 5)
df = pd.DataFrame(data, columns=['a', 'b', 'c', 'd', 'e'])

print(df.head())
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import BisectingKMeans

bisect_model = BisectingKMeans(n_clusters=4, random_state=42)
bisect_labels = bisect_model.fit_predict(df)

plt.figure(figsize=(8, 5))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=bisect_labels, cmap='viridis')
plt.title("Divisive Clustering Results (a vs b)")
plt.xlabel("a")
plt.ylabel("b")
plt.colorbar(label='Cluster ID')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df_scaled)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=bisect_labels, cmap='plasma', alpha=0.8, edgecolors='w')

plt.title("Divisive Clustering Visualized in 2D PCA Space (5D data compressed)")
plt.xlabel("Principal Component 1 (captures most variance)")
plt.ylabel("Principal Component 2")
plt.colorbar(scatter, label='Cluster ID')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
